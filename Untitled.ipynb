```python
mg = "URN:NBN:no-nb_digibok_2021120148028"
```

```python
dh.Models().models
```

```python
mg_ner = dh.NER(urn=mg, model='nb_core_news_lg', start_page=0, to_page=1000).ner
```

```python
mg_pos = dh.POS(urn=mg, model='nb_core_news_lg', start_page=0, to_page=1000).pos
```

```python
mg_pos[mg_pos.token == 'Isak']
```

```python
mg_pos[mg_pos.token == 'Inger']
```

```python
mg_group = mg_pos.groupby(['lemma', 'dep']).sum('frekv').reset_index()
```

```python
mg_ner[mg_ner.ner=="PER"].iloc[:10]
```

```python
mg_top = mg_ner[mg_ner.ner=="PER"].iloc[:10]['token'].values
```

```python
mg_cols = {x: dh.Collocations(corpus=[mg], words=[x], before = 2, after = 8).coll for x in mg_top}
```

```python
mg_forms = dh.WordLemma(list(mg_cols['Isak'].index)).lemmas
```

```python
mg_forms[mg_forms[0].str.contains('arbeid')]
```

```python
mg_forms[mg_forms[1] == 'verb'].groupby(2).count().sort_values(by=1, ascending=False).head(5)
```
